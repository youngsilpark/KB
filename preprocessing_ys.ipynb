# email

```python
import random
```

```python
string_pool = 'abcdedfhijklmnopqrstuvwxyz'
number_pool = '1234567890'
domain_pool = ['.com', '.kr', '.net', '.co.kr', '.ac.kr']
```

```python
def create_email(len_1, len_2):
      
  email = ''

  for l in range(len_1) :
    email += random.choice(string_pool+number_pool)

  email += '@'

  for l in range(len_2) :
    email += random.choice(string_pool)
    
  email += random.choice(domain_pool)

  print(email)
```

```python
for i in range(2500):
  create_email(random.randint(5,20),random.randint(5,10))
```

```python
CHO = (
    u'', u'', u'', u'', u'', u'', u'', u'', u'', u'',
    u'', u'', u'', u'', u'', u'', u'', u'', u''
)

JOONG = (
    u'', u'', u'', u'', u'', u'', u'', u'', u'', u'',
    u'', u'', u'', u'', u'', u'', u'', u'', u'', u'', u''
)

JONG = (
    u'', u'', u'', u'', u'', u'', u'', u'', u'', u'',
    u'', u'', u'', u'', u'', u'', u'', u'', u'', u'',
    u'', u'', u'', u'', u'', u'', u'', u''
)

JAMO = CHO + JOONG + JONG[1:]

NUM_CHO = 19
NUM_JOONG = 21
NUM_JONG = 28

FIRST_HANGUL_UNICODE = 0xAC00  # ''
LAST_HANGUL_UNICODE = 0xD7A3  # ''
#                      AB C DEFGHIJK L M N OPQR ST UVWXYZ
ENG_KOR_SUBSTITUENT = { 'B': '', 'C':'', 'K':'', 'L':'', 'M':'', 'N':'', 'R':'', 'T':''}

def decompose_index(code):
  jong = int(code % NUM_JONG)
  code /= NUM_JONG
  joong = int(code % NUM_JOONG)
  code /= NUM_JOONG
  cho = int(code)

  return cho, joong, jong

def hangul_index(letter):
    return ord(letter) - FIRST_HANGUL_UNICODE

def decompose(hangul_letter):
  """This function returns letters by decomposing the specified Hangul letter."""

  if hangul_letter in CHO:
      return hangul_letter, '', ''

  if hangul_letter in JOONG:
      return '', hangul_letter, ''

  if hangul_letter in JONG:
      return '', '', hangul_letter

  code = hangul_index(hangul_letter)
  cho, joong, jong = decompose_index(code)

  if cho < 0:
      cho = 0

  try:
      return CHO[cho], JOONG[joong], JONG[jong]
  except:
      print("%d / %d  / %d"%(cho, joong, jong))
      print("%s / %s " %( JOONG[joong].encode("utf8"), JONG[jong].encode('utf8')))
      raise Exception()
```
